## Training Model

### Splitting and normalize the data

For training:test data, we will split to 80:20, then split training to training and validation dataset.

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
N = 60

def create_sequences(data, n_steps):
    x, y = [], []
    for i in range(len(data) - n_steps):
        x.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(x), np.array(y)

X, y = create_sequences(scaled_data, N)
```

```python
temp_X, test_X, temp_y, test_y = train_test_split(X, y, test_size=0.2, shuffle=False)
train_X, val_X, train_y, val_y = train_test_split(temp_X, temp_y, test_size=0.2, shuffle=False)

train_X = train_X.reshape((train_X.shape[0], train_X.shape[1], 1))
val_X = val_X.reshape((val_X.shape[0], val_X.shape[1], 1))
test_X = test_X.reshape((test_X.shape[0], test_X.shape[1], 1))

print(f'Training set shape: {train_X.shape}')
print(f'Validation set shape: {val_X.shape}')
print(f'Test set shape: {test_X.shape}')
```

### Baseline RNN Implementation

```python

def build_rnn_model(hp):
    model = Sequential()
   
    for i in range(hp.Int('num_layers', 1, 4)):  
        model.add(SimpleRNN(units=hp.Int('units_' + str(i), min_value=32, max_value=256, step=32),
                            return_sequences=(i < hp.Int('num_layers', 1, 4) - 1),
                            input_shape=(train_X.shape[1], 1),
                            kernel_regularizer=l2(0.01)))
        model.add(Dropout(rate=hp.Float('dropout_' + str(i), min_value=0.1, max_value=0.5, step=0.1)))
    
    
    model.add(Dense(units=1))
    
   
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[0.1, 0.01, 0.001, 0.0001])),
                  loss='mean_squared_error')
    
    return model


tuner_rnn = kt.RandomSearch(
    build_rnn_model,
    objective='val_loss',
    max_trials=10,  
    executions_per_trial=1,
    directory='keras_tuner_dir',
    project_name='stock_price_prediction_rnn'
)

tuner_rnn.search(train_X, train_y, epochs=10, validation_data=(val_X, val_y), callbacks=[EarlyStopping(monitor='val_loss', patience=3)])

best_hps_rnn = tuner_rnn.get_best_hyperparameters(num_trials=1)[0]

rnn_model = tuner_rnn.hypermodel.build(best_hps_rnn)

rnn_history = rnn_model.fit(train_X, train_y, epochs=100, batch_size=best_hps_rnn.get('batch_size') if 'batch_size' in best_hps_rnn.values else 32, validation_data=(val_X, val_y), callbacks=[EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)])
plt.figure(figsize=(10, 5))
plt.plot(rnn_history.history['loss'], label='Simple RNN Training Loss')
plt.plot(rnn_history.history['val_loss'], label='Simple RNN Validation Loss')
plt.title('Simple RNN Model Loss During Training')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

rnn_predictions = rnn_model.predict(test_X)

rnn_predictions = scaler.inverse_transform(rnn_predictions)
true_values = scaler.inverse_transform(test_y.reshape(-1, 1))

rnn_predictions_shifted = rnn_predictions[:-1]
true_values_shifted = true_values[1:]

plt.figure(figsize=(14, 6))
plt.plot(true_values, color='blue', label='Actual Stock Price')
plt.plot(rnn_predictions, color='green', label='Simple RNN Predicted Stock Price')
plt.title('Stock Price Prediction - Simple RNN')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

rnn_mse = mean_squared_error(true_values, rnn_predictions)
rnn_mae = mean_absolute_error(true_values, rnn_predictions)
rnn_rmse = np.sqrt(rnn_mse)
rnn_mape = np.mean(np.abs((true_values - rnn_predictions) / true_values)) * 100
rnn_r2 = r2_score(true_values, rnn_predictions)

print(f'Simple RNN Mean Squared Error (MSE): {rnn_mse}')
print(f'Simple RNN Mean Absolute Error (MAE): {rnn_mae}')
print(f'Simple RNN Root Mean Squared Error (RMSE): {rnn_rmse}')
print(f'Simple RNN Mean Absolute Percentage Error (MAPE): {rnn_mape}%')
print(f'Simple RNN R^2 Score: {rnn_r2}')

plt.figure(figsize=(10, 5))
plt.plot(rnn_history.history['loss'], label='Simple RNN Training Loss')
plt.plot(rnn_history.history['val_loss'], label='Simple RNN Validation Loss')
plt.title('Simple RNN Training and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### LSTM Implementation

```python
def lstm_model(hp):    
    lstm_model = Sequential()
    for i in range(hp.Int('num_layers', 1, 4)):
        lstm_model.add(LSTM(units=hp.Int('units_' + str(i), min_value=32, max_value=256, step=32),
                       return_sequences=(i < hp.Int('num_layers', 1, 4) - 1),
                       input_shape=(train_X.shape[1], 1),
                       kernel_regularizer=l2(0.01)))
        lstm_model.add(Dropout(rate=hp.Float('dropout_' + str(i), min_value=0.1, max_value=0.5, step=0.1)))
    lstm_model.add(Dense(units=1))
    lstm_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[0.1, 0.01, 0.001, 0.0001])),
                  loss='mean_squared_error')
    
    return lstm_model

lstm_tuner = kt.RandomSearch(
    lstm_model,
    objective='val_loss',
    max_trials=10,
    executions_per_trial=1,
    directory='keras_tuner_dir',
    project_name='stock_price_prediction'
)

lstm_tuner.search(train_X, train_y, epochs=10, validation_data=(val_X, val_y), callbacks=[EarlyStopping(monitor='val_loss', patience=3)])

lstm_hps = lstm_tuner.get_best_hyperparameters(num_trials=1)[0]

lstm_model = lstm_tuner.hypermodel.build(lstm_hps)

lstm_history = lstm_model.fit(train_X, train_y, epochs=50, batch_size=lstm_hps.get('batch_size') if 'batch_size' in lstm_hps.values else 32, validation_data=(val_X, val_y), callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)])

plt.figure(figsize=(10, 5))
plt.plot(lstm_history.history['loss'], label='Training Loss')
plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

lstm_predictions = lstm_model.predict(test_X)

lstm_predictions = scaler.inverse_transform(lstm_predictions)
true_values = scaler.inverse_transform(test_y.reshape(-1, 1))

lstm_predictions_shifted = lstm_predictions[:-1]
true_values_shifted = true_values[1:]

plt.figure(figsize=(14, 6))
plt.plot(true_values, color='blue', label='Actual Stock Price')
plt.plot(lstm_predictions, color='red', label='Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

mse = mean_squared_error(true_values, lstm_predictions)
mae = mean_absolute_error(true_values, lstm_predictions)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((true_values - lstm_predictions) / true_values)) * 100
r2 = r2_score(true_values, lstm_predictions)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Percentage Error (MAPE): {mape}%')
print(f'R^2 Score: {r2}')

plt.figure(figsize=(10, 5))
plt.plot(lstm_history.history['loss'], label='Training Loss')
plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
